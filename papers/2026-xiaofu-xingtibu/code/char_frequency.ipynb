```python
# ===== 파일 불러오기 + 글자 세기 =====
import re

# (1) 목표 = 10.txt(형체부), 참조 = 나머지 12권
TARGET_FILE = "10.txt"
REF_FILES = [f"{i:02d}.txt" for i in range(1, 14) if i != 10]

# (2) 파일 하나를 열어 글자를 통째로 읽어오는 함수
def load(path):
    with open(path, encoding="utf-8") as f:
        return f.read()

# (3) 읽어온 글자에서 한자만 골라내는 함수 (𩯽 같은 희귀자 포함)
CJK = re.compile(r'[\u3400-\u4dbf\u4e00-\u9fff\uf900-\ufaff\U00020000-\U0002ebef]')
def tokenize(text):
    return CJK.findall(text)

# (4) 실제로 읽어서 한자만 추출
target = tokenize(load(TARGET_FILE))
ref = tokenize("".join(load(p) for p in REF_FILES))

# (5) 글자 수와 종류 수 출력
N1, N2 = len(target), len(ref)
print(f"목표(형체부) N1 = {N1:,}자 / 유형 {len(set(target)):,}종")
print(f"참조(12부)   N2 = {N2:,}자 / 유형 {len(set(ref)):,}종")
```

```python
# ===== 3단계: 두 코퍼스 빈도를 하나의 표로 합치기 =====
import pandas as pd
from collections import Counter

# (1) 각 코퍼스에서 글자별 출현 횟수를 센다
c_target = Counter(target)   # 형체부: {글자: 횟수}
c_ref    = Counter(ref)      # 참조 12부: {글자: 횟수}

# (2) 두 코퍼스에 등장한 모든 글자를 모은다(합집합)
all_chars = set(c_target) | set(c_ref)

# (3) 글자마다 [형체부 횟수 a, 참조 횟수 b]를 짝지어 표로 만든다
df = pd.DataFrame({
    "char": list(all_chars),
    "a": [c_target.get(ch, 0) for ch in all_chars],  # 형체부 출현 수
    "b": [c_ref.get(ch, 0)    for ch in all_chars],  # 참조 출현 수
})

# (4) 형체부에 많이 나온 순으로 정렬
df = df.sort_values("a", ascending=False).reset_index(drop=True)

# (5) 표 크기와 윗부분 확인
print(f"표 크기: 글자 {len(df):,}종")
print(df.head(10))
```
